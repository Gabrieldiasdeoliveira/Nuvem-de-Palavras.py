# Nuvem de Palavras

O trabalho realizado consistiu na geração de uma nuvem de palavras utilizando a linguagem de programação Python. A nuvem de palavras é uma representação visual de um conjunto de palavras em que o tamanho de cada palavra é proporcional à sua frequência no texto original.

Para criar a nuvem de palavras, foram utilizadas as seguintes etapas:

>1. Pré-processamento dos dados: Os dados de entrada, que representavam um mtexto retirado do blog do DataSeince Academy, foram preparados e processados para remover elementos indesejados, como pontuação, caracteres especiais e stopwords (palavras comuns que não trazem significado para a análise).

> 2. Tokenização: O texto dos leads foi dividido em palavras individuais, chamadas de tokens, para que pudessem ser processadas separadamente.

> 3. Contagem de frequência: Foi realizada a contagem da frequência de cada palavra.

> 4. Geração da nuvem de palavras: Utilizando a biblioteca "WordCloud" do Python, foi criada a nuvem de palavras com base na frequência das palavras. As palavras mais frequentes receberam tamanho maior e foram posicionadas de forma aleatória na nuvem.

> 5. Personalização da nuvem de palavras: Foram aplicadas configurações adicionais para melhorar a aparência da nuvem de palavras, como escolha de cores, ajuste do tamanho das fontes e definição de formato da nuvem.

> O resultado final foi uma nuvem de palavras que representou visualmente as principais palavras presentes no texto forncido pela empresa, destacando as mais frequentes. Essa representação visual facilita a identificação de termos-chave e temas relevantes relacionados ao assunto.

# Plot da Nuvem de Palavras

Nessa etapa eu utilizei algumas opções de nuvens, mas dentro da biblioteca existem varias outras opções de cores.

```python
# Nuvem sem Máscara
wordcloud = WordCloud(background_color='white',
                      stopwords=stopwords_customizada,
                      contour_width=1,
                      contour_color='white',
                      colormap='pink',
                      random_state=42)

wordcloud.generate(texto_filtrado)
```

```python
# Plotar a nuvem de palavras
plt.figure(figsize=(8,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
